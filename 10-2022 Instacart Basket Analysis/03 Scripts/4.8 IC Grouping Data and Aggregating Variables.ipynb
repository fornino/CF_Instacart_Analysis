## Loyalty flag column

```python
# Creating max_order with transform()
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
pd.options.display.max_rows = None
```

```python
# Check output
ords_prods_merge.head(100)
```

```python
# dropping extra column
ords_prods_merge = ords_prods_merge.drop(['Unnamed: 0'], axis = 1)
```

```python
# Creating loyalty flag column
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check output
ords_prods_merge.head(50)
```

```python
ords_prods_merge['loyalty_flag'].value_counts()
```

## Spending flag column

```python
# Creating average_order with transform()
ords_prods_merge['average_order'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check output
ords_prods_merge.head(50)
```

```python
# Creating spending flag column
ords_prods_merge.loc[ords_prods_merge['average_order'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_order'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Check output
ords_prods_merge.head(50)
```

```python
ords_prods_merge['spending_flag'].value_counts()
```

## Frequency flag column

```python
# Creating order_frequency with transform()
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check output
ords_prods_merge.head(50)
```

```python
# Creating frequency flag column
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <=20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check output
ords_prods_merge.head(50)
```

```python
ords_prods_merge['frequency_flag'].value_counts()
```
