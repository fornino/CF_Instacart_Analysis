# 03. Data Wrangling

# 04. Data Consistency Checks

## Mixed-Type Data

```python
# Checking for any mixed-type data
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len(customers[weird]) > 0:
      print(col)
```

```python
# Checking for missing values to see if that is the cause of mix
customers['first_name'].isnull().sum()
```

```python
customers[customers['first_name'].isnull() == True]
```

#### There are 11259 missing values for first name. We are still able to identify customers with the user_id key so the missing values will remain NaN

```python
# Making all values in first_name strings
customers['first_name'] = customers['first_name'].astype('str')
```

## Missing Data

```python
# Checking for missing data in rest of df
customers.isnull().sum()
```

#### There is no other missing data

# 05. Combining customers with ords_prods

```python
# Import ords_prods_merge'
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merged_updated2.pkl'))
```

```python
ords_prods_merge.head()
```

```python
# Dropping _merge column from previous merge
ords_prods_merge = ords_prods_merge.drop('_merge', axis = 1)
```

```python
ords_prods_merge.shape
```

```python
# Checking datatype of dfs to ensure they match
ords_prods_merge.dtypes
```

```python
ords_prods_merge['user_id'] = ords_prods_merge['user_id'].astype('str')
```

```python
# Testing merge
pd.merge(ords_prods_merge, customers, on = 'user_id', indicator = True)
```

```python
ords_prods_custs = ords_prods_merge.merge(customers, on = 'user_id', indicator = True)
```
